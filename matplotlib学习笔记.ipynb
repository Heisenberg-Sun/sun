```python
import matplotlib as mpl  
import matplotlib.pyplot as plt

#表示创建一张图1,这个1相当于图的ID. # 画图之前首先设置figure对象，
#此函数相当于设置一块自定义大小的画布，使得后面的图形输出在这块规定了大小的画布上，其中参数figsize设置画布大小
fig=plt.figure(1)  #控制dpi、边界颜色、图形大小、和子区( subplot)设置.等价于plt.title(1)
plt.plot([1,2,3,4])
fig=plt.figure(2)  #控制dpi、边界颜色、图形大小、和子区( subplot)设置
plt.plot([1,2,3,5])
```

```python
fig=plt.figure(figsize=(12,8)) #设置画图板。
fig.set(alpha=0.2)#透明度设置

plt.subplot2grid((2,3),(0,0)) #画子图，分成2×3个格子，当前图的开始坐标是（0，0）
plt.plot([1,2,3])
plt.title('sample') #图的标题
plt.ylabel('people')#y轴标签
plt.xlabel('age')#x轴标签
plt.legend(['age/people'],loc='best') #添加图例，一定要有loc，不然就会报错.前面的字符或者字符串一定要加[]或者（）

plt.subplot2grid((2,3),(0,1))
plt.title('scatter')
plt.scatter([1,2,3,4,5],[1,1,3,4,5])#绘制散点图
plt.grid(b=True) #显示网格
```

## 各种图的画法

```python

# 1D data
x = [1,2,3,4,5]
y = [2.3,3.4,1.2,6.6,7.0]

plt.figure(figsize=(12,6))

plt.subplot(231)
plt.plot(x,y) #线型
plt.grid()#画出网格
plt.title("plot")

plt.subplot(232)
plt.scatter(x, y)#散点图
plt.title("scatter")

plt.subplot(233)
plt.pie(y)  #饼型图
plt.title("pie")

plt.subplot(234)
plt.bar(x, y) #条形图
plt.title("bar")
```

```python
import pandas as pd
#对于pandas数据来说，可以直接使用plot函数，方便的绘制条形图
d1=pd.Series([3,4],index=[0,1])
d2=pd.Series([1,5],index=[0,1])

fig=plt.figure(figsize=(8,6))
fig.set(alpha=0.7)
plt.subplot(121)#画子图，与subplot2grid方式一致
d1.plot(kind='bar')
plt.ylim(0.0,8)
plt.xticks([0,1],['true','false'])#给对应的x刻度设置标签

plt.subplot(122)
df=pd.DataFrame({'d1':d1,'d2':d2})
df.plot(kind='bar',stacked=True)#条形图，层叠的方式显示
plt.show()
```

```python
fig=plt.figure(figsize=(12,6))
fig.set(alpha=0.65)
#画子图的第三种方式
ax1=fig.add_subplot(141)#返回的就是这个子图
d1.plot(kind='bar')
ax2=fig.add_subplot(142,sharey=ax1)#表示这个新的子图与ax1共享y轴坐标
plt.grid()
d2.plot(kind='bar')
plt.show()
```

```python
plt.fill_between(x,y1,y2,alpha=0.1,color='b')
#以x的坐标给y1-y2两条线之间的区域填充颜色，alpha设置透明度，color设置颜色
plt.plot(x,y,'o-',color='r',label='line')#画出一条线，线型选择为‘o-'，颜色为red,线的图例为'line'，下面需要指定一个图例的位置
plt.legend(loc='best')
```
